```python
#welcome
name = input("Enter your name: ")
print("What's up %s, want to play hangman?" % name)
answer = input("Yes or No: ")
game = 0
while game == 0:
    if answer == "Yes":
        break
    if answer == "yes":
        break
    else:
        print("See you later")
print("")

#countdown to play
import time
time.sleep(0.1)
print("Loading...")
time.sleep(1)
print("3")
time.sleep(1)
print("2")
time.sleep(1)
print("1")
time.sleep(1)
print("")
print("Let's play!")

#game
import random
fruit = ("pineapple", "starfruit", "papaya", "dragonfruit", "jackfruit",  "lychee")
randomwords = ("xylophone","atrocious","spectacular")
languages = ("french","vietnamese","english","chinese","hindu","korean")

alist = (fruit,randomwords,languages)
aalist = random.choice(alist)
word = random.choice(aalist)

print("")
time.sleep(0.2)

if aalist == fruit:
    print("Your topic is: fruit")
elif aalist == randomwords:
    print("Your topic includes random words, good luck :D")
elif aalist == languages:
    print("Your topic is: languages(in lower case)")
    
guesses = ''
turns = 6 #head,body,legs,arms

#gameplay
while turns > 0:         
    fail = 0             
    for char in word:      
        if char in guesses:    
            print(char)    
        else:
            print("_")     
            fail += 1    
    if fail == 0:        
        print("Congratulations %s, you won!" % name)
        break              
    guess = input("Guess a character:") 
    guesses += guess
    
    if guess not in word:  
        turns -= 1        
        print("Wrong")  
        print("You have %d more guesses" % turns)
        if turns == 0:           
            print("You have lost, try again!")  
```
